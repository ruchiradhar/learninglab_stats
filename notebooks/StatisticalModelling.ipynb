# What is Modelling?

## Basics of Modelling

**Definition**: The process of modelling refers to fitting models (hypothesized based on our background knowledge) to data (about the phenomenon we're interested in). 

**Purpose**: There can be three possible types of knowledge modelling gives us

1. Distributional information of variables (maybe conditioned on other variables)
2. Summary information on relationships between variables
3. Prediction of values of variables based on other variables

**Types of Models**: There are two main types of models

1. **Parametric Models**: The relationship between entities is assumed to follow a specific form with a fixed number of parameters that fully describe the model. Eg: Linear Regression, Logistic Regression, Normal Distribution (with mean $\mu$ and variance $\sigma^2$)
2. **Non-parametric Models**: The relationship between entitiesis not assumed to have any specific form. Eg: Decision Trees, K-Nearest Neighbours etc.

The former is typically simpler and more interpretable while the latter can handle complex relationships. 

**Steps in Modelling**: 

1. Data distribution checkf or predictors
2. Fit model by estimating parameters
3. Assess model fit
4. Improving model fit (explain remaining variability by including more predictors)

Note: Parameters vs Predictor

## Process of Modelling


**Dependent vs Independent Variables**

The dependent variables (a.ka endogenous variables) are those variables of interest for which we want to model distributional features as a function of certain other independent variables (a.ka exogenous variables). The distributions of dependent variables are assumed to depend on the values of independent variables. 

*Eg: If we try to estimate how age affects income, age is an independent variable while income is the dependent variable.* 

**Experimental vs Observed Variables**

The values of variables maybe of two types depending on how they have been obtained: 

1. Experimental Variables: The values are obtained as a result of randomized experiments manipulated by investigators.
2. Observed Variables: The values are simply observed as occuring from a natural process. It is more difficult to make causal inference for such variables. 

**Modelling Variables**

The goal of a modelling exercise is to "model" the DV, a variable of primary interest, as a function of other theoretically relevant IVs. The process is as follows: 

1. We define a research question with both DVs and IVs.
2. We determine the types of all variables i.e continuous/categorical/binary etc. If continuous, estimate functional relationship of IV with distributional features of DV. If categorical, compare groups defined by categories in terms of distributions of DV.
3. We select a reasonable distribution of DV (eg: normal) and defining parameters of that distribution (eg: mean,variance).
4. We analyse the distribution of DV as a function of IVs.
   
*Eg: We assume blood pressure is normally distributed continuous variable where mean depends on IVs age, body-mass index and gender.*

**Precautions**

*Missing Data*: In real world data, there can often be missing values. 

1. Conduct descriptive and bivariate analyses for both DVs and IVs.
2. Resolve missing data by either deleting any rows of data with missing values for either DV or IVs or replacing missing values with mean.
3. Compare units with missing data vs units with complete data to ensure they are not systematically different.

*Controlling Confounding Variables*: For a given dependent variable, we might have some other relevant "confounding variables" other than our theorized IVs which can have negative impact on our estimation of IV-DV relationships. As a result, we need to control for these by: 

1. Randomizing Study Designs: The randomized groups should be balanced with respect to confounding variables.
2. Analyzing Non-Randomized Designs: Any observational studies should be analyzed for where groups are not balanced.
3. Adding IVs: Add any possible confounding variables as independent variables to adjust for the confounding effect.
