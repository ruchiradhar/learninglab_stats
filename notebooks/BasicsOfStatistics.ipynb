# What is Statistics?

**Definition:** Statistics is the study of the tools and methods that allow us to learn from data. It involves: 

1. Measuring and collecting data in appropriate manner
2. Understanding new types of data and statistical methods for their analysis
3. Understanding properties of statistical methods 
4. Development of new statistical methods

**Goal**: Assessing whether claims based on data are meaningful and how meaningful/close to truth they are.


**Historical Milestones**: 

1.  **Ancient Times**: It gave rise to data collection on populations, harvests, and natural calamities.
2.  **18th Century**: It gave rise to probability theory as the study of randomness and variation.
3.  **19th Century**: It gave birth to modern statistics via demographical studies and genetics.
4.  **20th Century**: It led to significant advances in statistical theory and statistical learning with rise of computers.
5.  **21st Century**: It has led to huge data stores (Big Data) and advanced applications of statistical methods on data (Machine Learning).


**Note: *Statistics* is not the same as a *statistic* which is a numerical/graphical summary of a collection of data.**

# Study Design

It refers to the different types of research studies that give rise to data. It is crucial for two things: 

1. **Power Analysis**:  The process to assess whether a study design is likely to yield meaningful findings
2. **Bias Analysis**:  The process to determine whether  sample is not representative of population or if measurements are systematically off target. 

## Confirmatory vs Exploratory

1. Confirmatory Studies employ the scientific method (Falsifiable Hypothesis -> Data Collection -> Testing Hypothesis Validity)

2. Exploratory Studies collect and analyze data without pre-specifying question. 

While the latter is informative, it can be misleading. The more questions we ask from a dataset, the more likely we are to draw a misleading conclusion (multiple testing, p-hacking, overfitting). 


## Comparative vs Non-Comparative

1. Comparative Studies compares a feature between two things. Eg: Harvest of Oranges in Spain vs Italy
2. Non comparative Studies estimate the value of a feature. Eg: Stock Price prediction.


## Observational vs Experimental

1. Observational studies give rise to natural grouping of data. Eg: Lifespan of Smokers vs Non Smokers
2. Experimental studies involve manipulation of different groups. Eg: Drug Trials

The observational studies are often susceptible to bias where sample is not representative or measurements are systematically off-target.

# Sampling Design

In Statistics, sampling forms a very integral part of the experiment process.  In reality, we only get to work with one random sample of from the population and need to estimate statistics for the population from this sample. The most common way to do this is to use random sampling. Random sampling allows us to estimate features of the sampling distribution that would emerge if we selected multiple random samples from the population. 

Random sampling is a scientific tool for making inferences about larger populations. With a representative, well-designed random sample of units from a well-defined target population of interest, we can make sound scientific conclusions about population features of interest (e.g., mean income), and we don’t need a very large sample to do so. 

## Populations and Samples

Up until early 20th century, researchers attempted to take measurements of every unit in a population (i.e census). It was only in 1930s, that [Jerzy Neyman's work](https://www.jstor.org/stable/2342192) enabled use of samples and random sampling.

### Defining population

Who is the target set of people? What time frame? Where is population located?. It should be clearly defined

### Types of Sampling

**Option 1**: Conduct census (suitable for small populations). Eg: Population Census of countries

**Option 2**: Select scientific probability sample where units probability of selection is known. 

**Option 3**: Select non-probability sample where units probability of selection is not known. 

#### Type 1: Probability Sampling

It involves: 

1. Construct list of all units in population (sampling frame)
2. Determine probability of selection of every unit in frame (should be known and non-zero)
3. Select unit from list at random (based on different subgroups with different probabilities of selection)
4. Measure units in this random sample

Why probability sampling?

1. The known probabilities of selection reduces risk of bias during sample selection
2. It allows good estimates of population features based on sampling distributions
3. It allows uncertainty estimation of population features

Types of probability sampling:

1. **Simple Random Sampling (SRS)**: Given $N$ population units, select $n$ sample units where all have equal probability of selection ($n/N$) and all possible samples of size $n$ are equally likely. This gives IID data and makes feature estimates unbiased and closed to population values. It can be: with replacement or without replacement.

2. **Complex Random Sampling**: Each unit has a known probability of selection but have more complicated design features like stratification (ensures representation of different subgroups) and cluster sampling (ensures variety akin to population within strata). A unit's probability of selection depends on: clusters sampled from each stratum ($a$), total clusters in each stratum ($A$), units sampled from each cluster ($b$), total units in each cluster ($B$) as $P (x)= (a/A)(b/B)$. The inverse of $P(x)$ is the sampling weight i.e if $P(x)=1/100$, then sampling weight of $x$ is 100 i.e they represent x and 99 others in the population (useful for unbiased population estimates). 

Thus, good sampling methods give realistic and representative of samples.

#### Type 2: Non-Probability Sampling

It involves selection of units such that probability of selection cannot be determined and there is no random selection. This creates a risk of sampling bias and lower validity of population estimates. It can be improved with: 

1. **Pseudo-randomization:** The non-probability sample is combined with a probability sample that contains similar measurements. Use logistic regression to estimate the probability of being included of units in NPS using auxiliary information and use estimated probabilities for valid population estimates. 

2. **Calibration:** The weights of responding units in NPS is computed to allow weighted sampled to mirror a known population. Eg: If NPS has 70% female and 30% male while population has 50% each then we down-weight females and up-weight males.

For more details refer to [Elliott & Valliant (2017)](https://www.jstor.org/stable/26408228?seq=1).

Example: Volunteer Sampling (Opt-In Survey), Quota Sampling (Target-Based Selection), Snowball Sampling (Referral Selection), Convenience Survey (Twitter Data)

## Sampling Distribution

A sampling distribution is a key part of statistical inferences. 

### Why do we need a Sampling Distribution?

Let's understand the need for sampling distributions this step-by-step, with an example population and sample: 

* A distribution refers to the values on a variable of interest (Eg: The distribution of SAT scores in Kentucky in 2012)
* For the target population (SAT Examinees in Kentucky in 2012), the SAT scores would follow a certain distribution (Normal Distribution) and population parameter would have a certain value (Eg: Population Mean SAT Score)
* For a probability sample of the population, the SAT scores would also follow a certain distribution and the survey estimated mean would have a certain value (Eg: Sample Mean SAT Score).
* To make valid inferential statements about the population from the sample, we would like the estimated parameter (Sample Mean SAT Score) to be close to the population parameter (Population Mean SAT Score).  This is formalized with the help of a sampling distribution.

### What is a Sampling Distribution?

A sampling distribution refers to the distribution of survey estimated parameters we would see if we had selected many random samples (using same sample design as our actual sample)  and computed an estimated parameter from each of those samples. Some key properties of sampling distributions: 

1. It is hypothetical since in practice we cannot afford to draw thousands of such samples.
2. The sampling distribution can have different nature than the variable distribution. Eg: The distribution of SAT Scores in Kentucky can look different from the distribution of means of SAT Scores across random samples from Kentucky.
3. The sampling variance (variability of the estimated parameter across hypothetical samples) comes from sampling error (the difference in sample estimated parameter and population parameter for a given sample).
4. The sampling distribution would have zero sampling variance and zero sampling error if samples are large enough since all our sample estimates would equal the population parameter. 

The sampling distribution changes with different sampling methods. Eg: Cluster Sampling (with less clusters and more sampling within a cluster) has been shown to result in higher sampling variance compared to Simple Random Sampling (SRS). Ideally, our sampling distribution should have low sampling variance such that our estimated parameter comes close to the population parameter. 


### Central Limit Theorem

The Central Limit Theorem is closely linked to the idea of sampling distributions. It states that: 

*"If we take sufficiently large and IID samples from a population, the distribution of sample estimates (the sampling distribution) will be approximately normally distributed"*

**Explanation of Idea:** If we repeat the sampling process many times, we come closer to the objective (population) value of the parameter. So, the proportion of times a value occurs across the infinite sampling process brings us closer to the true probability of the parameter having that value.  This also gives rise to confidence intervals (CI) where a 95% CI indicates that across infinite sampling, 95% of times that interval would contain the true mean. 

**Philosophy of Idea:** The idea behind the CLT represents the Frequentist Philosophy of probability as the long-run frequency of outcomes and uncertainty comes from the sampling distribution.  The alternative Bayesian Philosophy view probability as a degree of belief in the value of a parameter and the uncertainty comes from the posterior distribution. Thus, CLT is only used as a justification tool in Bayesian Probability. 

**Exceptions:** When we have non-normal sampling distributions, we need more specialized procedures to make population inferences. This is where Bayesian methods can help us since we can no longer depend on the CLT to make inferences about the population.

### Design Effect

Source:[Coursera Statistic with Python](https://www.coursera.org/learn/understanding-visualization-data/supplement/TY3qn/cluster-sampling-and-design-effects)

The design effect (DEFF) is the ratio of the sampling variance of a given estimate under a complex probability sampling design to the sampling variance of a given estimate under a simple random sample (without replacement) of the same size:

$$DEFF = \frac{var(est)_{Complex} }{var(est)_ {SRS}}$$

If a given complex probability sampling design has a DEFF greater than 1, this means that the complex sample design is increasing the variance of a given estimate while still producing the same sample size. Leslie Kish, in his classic book Survey Sampling, showed that the DEFF can be written as follows for estimates of means based on cluster samples:

$$DEFF = 1 + (b – 1) × \rho$$

where $b$ is the average sample size selected from each cluster, and $\rho$ is the intra-cluster correlation of the variable used to compute the mean (i.e., the correlation of any two values on the variable of interest for elements coming from the same cluster). So, the DEFF will become larger if:

1. The average sample size selected per cluster, b, becomes larger (just like we saw in the simulation study!), and/or

2. The intra-cluster correlation, which is typically larger than 0 for clustered human populations, is larger than zero.

Putting this all together, we see that the design decision to sample fewer clusters and sample more elements per cluster (which has the potential to save money, because you aren’t visiting as many clusters to collect data) will inflate the sampling variance, unless the intra-cluster correlation is zero (which is again unlikely for clustered populations). A key observation here is that DEFF is not a function of the number of clusters sampled, but rather how many elements are sampled from each cluster on average.

## Sampling Theory

Sampling Theory, developed by Jerzey Neyman in 1930s, forms the bridge between our actual probability sample and the hypothetical sampling distribution. 

* Given the CLT applies, it gives us methods to estimate features of the sampling distribution from our actual sample.
* With only one probability sample, we can know about the expected sampling distribution and sampling variance.
* With the sampling variance, we can know where most estimates will fall in the distribution.
* With knowledge of where most estimates will fall, we can infer probability and uncertainty of the values of the population parameter. 

# Inference Design

As we discussed, the way we sample has significant impacts on how we make inferences about the population from our sample. 

1. With Probability Samples: Under the normality assumption from Central Limit Theorem (CLT), there are two approaches for making population inferences: Confidence Interval Estimation and Hypothesis Testing. If not normal, alternative approaches like modelling or data combinations are used.

2. With Non Probability Samples: We have to use other methods of inference.

   
We will take a look at how each of these approaches work.

## For Probability Samples

### Confidence Interval Estimation

**Step 1**: Compute unbiased point estimate of parameter of interest from sample (average of all possible values for point estimates across samples). The sampling distribution centers on the true population parameter value. If sample points had unequal probabilities of selection, account for those in this computation.

**Step 2**: Compute an unbiased estimate of the sampling variance (from the hypothetical sampling distribution under the sample design). The square root of the variance gives us the Standard Error (the standard deviation of the point estimate).

**Step 3**: Form a confidence interval (CI) as *Best Estimate $\pm$ Margin of Error* where *Best Estimate* is the unbiased point estimate and *Margin of Error* is a few estimated Standard Errors (where few depends on desired confidence level and sample design). 

**Step 4**: Form an interpretation at given confidence level. Eg: 95% confidence level translates to 95% of intervals constructed this way having true parameter in them or that we have 0.05 significance.

*Note: If the sample design (via survey weights) is accounted for when computing the estimate of the parameter of interest, the stimate maybe biased and the confidence interval maybe shifted away from the true population parameter.*

### Hypothesis Testing

**Step 1**: State hypothesis as *The value of our parameter is x* where $x$ is the null value. The question we are asking is *Is the point estimate for parameter close to the null value or far away*?

**Step 2**: Compute Test Statistic $= \frac{estimate-x}{standard error}$.  

**Step 3**: Consider if $x$ is correct i.e if null is true, what is the probability of seeing a test statistic this or more extreme? If probability is small, reject the null hypothesis.

### Complex Sampling

For complex samples, we do stratification (we allocate overall sample to different strata with different allocation methods) and clustering (we do random sampling of clusters of population units across multiple stages) with weighting (certain sungroups sampled at higher artes creating unequal probabilities of selection). For making inferences on parameters, the goal is to minimize sampling variance for particular variables. This can be done with following schemes: 

1. **Proportionate Allocation**: Following actual population distribution proportions for different strata. This reduces between-stratum variance in means or other relevant parameters, thsu reducing overall sampling variance.
2. **Clustering Analysis**: Accounting for clustering effects like high correlation of units from a cluster while estimating sample variance.
3. **Weighting Analysis**: Accounting for the unequal probabilities of selection by weighting and rates of response ($Weight=1/Probability * 1/ResponsePercentage$) during estimating sample variance. However, highly variable survey weights will lead to high sampling variance.
4. **Visualizing Design Effects**: It is useful to keep in mind how such things effect the sampling variance.

## For Non-Probability Samples

### Quasi-Randomization

This combines data from non probability sample and a new probability sample. 

1. Stack the two samples together and track which points are member of non probability sample vs probability sample.
2. Fit logistic regression model to predict whether a point is the non probability sample, using the common variables of the samples as model features. We weight non-probability cases by 1 and probability cases with survey weights.
3. For the points in non-probability sample, we get survey weights from predicted probabilities as $SurveyWeights= \frac{1}{PredictedProbability}$
4. Use survey weights for calculating probability of sampling.

Problem: It is difficult to estimate sample variance. We can simulate sampling distribution by bootstrap samples or jackknife samples.

### Population Modelling

1. Use predictive modelling to predict sample totals on key parameters of interest for population units not in the non-probability sample.
2. Compute estimates of interest using estimated totals. Eg: $WeightedMean= \frac{PredictedTotalEstimate}{Estimated Population Size}$

Problem: We need good predictive models. Standard errors also need to be modelled or computed from resampling.

# Data Design

It refers to the considerations around the source, type and management of data.

## Sources of Data

It can take on many forms: numbers, images, text, audio, etc. But where does data come from?

1. **Organic/Process Data**: It refers to data that are generated as a result of a process (POS, Stock Market, Search History, etc). It often leads to Big Data that can be "mined" with statistical methods to study trends and relationships.
2. **Designed Data**: It refers to data collected to address a specific research objective via sampling (Tweet Extraction, Census Survey). It is often smaller than organic data and do not reflect natural trends.

## Type of Data (IID)

The data source has direct implications for an important consideration: **IID Data**. This refers to data being independent (each data point does not influence others) and identically distributed (all data points come from the same probability distribution). It is important because: 

1. Under independence assumptions, probabilities can be multiplied easily.
2. Under identical distributions, we can summarize with a single statistic.
3. Many statistical methods can only apply to IID data (Central Limit Theorem, T Test, Regression etc)

For Non IID Data: 

1. If Dependent: We need to use methods/models that account for autocorrelation in data (like time-series/clustered models)
2. If Heterogeneous: We need to model data from different distributions separately or include covariates (like mixed effects models)

## Data Management

It refers to all steps of data processing that occur after data is collected but before data analysis. Some key principles: 

1. Never modify source data files.
2. Use scripts to load source data in and generate analysis from data.
3. In scripts, use appropriate variable names and analyze missing values.

Some common formats of data storage: 

1. **Rectangular Data:** The rows represent individual units from population and columns represent variables for each unit.
2. **Repeated Measures Data:** There are multiple measurements made on each unit in a study and can have two forms: wide format (one row per subject) and long format (one row per measurement). The former is more convenient when each unit has same number of measurements while latter is more convenient for regression analysis. 

Some common tools for data management include: 

1. **Database Software (e.g. SQL):**  They can be very useful for large-scale data management. Some statistical software can read data directly from a database.  Another approach is to construct a text data file from a database e.g. using SQL.
2. **Big Data Software (e.g. Hadoop and Spark):** They can be useful for manipulating very large datasets.
3. **Binary Data Software (e.g. HDF5, Apache Parquet,Apache Arrow):** These are open-source standards for large binary datasets.  Using these formats saves processing time relative to text/csv because fewer conversions are performed when reading and writing the data.
4. **Compression Software (e.g. Gzip):** These allow saving large data sets in compressed form and reading them into statistical software directly from the compressed file.  This allows the data to be read much faster, and reduces storage space.

### Scipy

SciPy provides a large number of functions that are useful for different types of scientific and engineering applications.  Many SciPy functions operate on numpy arrays. The most useful is `scipy.stats` which calculates some important functions for many probability distributions:

* **Probability Mass Function (PMF)**: It gives the probability of a random variable $X$ at point $x$. For any discrete random variable $X$, $PMF(x)= P(X=x)$. Eg: PMF (3)= 1/6 where X= Die Roll Outcome
* **Probability Density Function (PDF)**: It gives the height of the curve $PDF(X)$ at point $x$. The integration of PDF gives the probability. For any continuous random variable, $PDF(a<X<b)$= $\int_a^b f(x) dx$. 
* **Cumulative Distribution Function (CDF)**: The total probability $P(X=x)$ up until point $x$ which is either cumulative sum of PMF (for discrete variables) or total area under PDF (for continuous variables).
* **Percent Point Function (PPF)**: It is the inverse of the CDF and given a probability $q$, calculates the value of x at that point. It gives the value $x$ such that $P(X ≤ x) = q$

```python
# importing
from scipy import stats

# PMF
# The probability of getting exactly 3 heads in 10 flips of a fair coin
print(stats.binom(10, 0.5).pmf(3))

# PDF
# The height of the standard normal density function at 1
print(stats.norm.pdf(1))

# CDF
## CDF of Normal Distribution at zero (the median)
print(stats.norm.cdf(0))

# CDF of Binomial Distribution at less than or equal to 3
# The probability of getting 3 or fewer heads in 10 flips of a fair coin
print(stats.binom(10, 0.5).cdf(3))

# CDF of Exponential Distribution at less than or equal to 3
#The probability that a standard normal value is less than or equal to 3
print(stats.expon.cdf(3))

# PPF
# The median of a standard Student's t distribution with 10 degrees of freedom
print(stats.t(10).ppf(0.5))
```

### Univariate Analysis

Some common ways to summarize quantitative data for a single variable are: 

1. **Measures of Central Tendency**: A numerical summary to indicate the central and most important values of a variable, around which most of the data lies. There are three main measures: mean (average value), median (middle value), and mode (most frequent value).
   
2. **Measures of Spread**: A numerical summary to indicate how far data points are from the central values. There are four main measures: range (the difference between max and min values), IQR (the diff of 3rd Quartile and 1st Quartile i.e the range between which 75% and 25% of the data lie), variance (the average distance of a data point from mean), and standard deviation (the square root of variance).

3. **Histogram**: A graphical summary where we divide the value range into bins and observe frequency of each bin in the dataset. The main aspects of a histogram are shape (symmetric, bell, skewed, etc), modality (how many peaks indicating different distributions), center (mean, median), spread (range, IQR, variance, etc), and outliers (datapoints far away from bulk).

4. **Boxplot**: A graphical summary where the dots give outliers, whiskers are $Point \pm 1.5 * IQR$, edges of a box give the Q3 and Q1, and the middle of the box lies at the Median. The axis of the box orientation indicates the values of the variable. 

Here are some important empirical facts to keep in mind: 

1. **Robust Estimator**: The median is a robust estimate of central tendency unlike the mean which is influenced by outliers. For example, in skewed distributions, the mean gets pulled more towards direction of the skew.

2. **Empirical Rule**: In a normal distribution, 68% of data are within 1 SD of the mean and 95% of data are within 2 SD of the mean and 99.7% of the data are within 3 SD of the mean.

3. **Z-Score**: A z-score or standard score indicates how unusual a data point is in a given distribution. A positive value indicates over the mean whereas negative value indicates below the mean. The formula is $\frac{Value-Mean}{SD}$

### Multivariate Analysis

Some common ways to summarize relationships between multiple categorical variables are: 

1. **Contingency Tables**: A frequency table where rows represent values of one variable and columns represent values of another variable. Each amount can be converted to percentages for more clarity.
2. **Comparative Bar Charts**: We have two bar charts, for two different variables. Alternatively, we can have side by side bar charts (each bar is for a value of the 2nd variable) or stacked bar charts (one bar for each variable 2 and sections within a bar for variable 1).
3. **Mosaic Plots**: They are directly representative of the frequency table where size of each cell is proportional to amount. Uneven lines in mosaic plots indicate dependent variation.

```python
# contingency table
# import
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats

# load data
da = pd.read_csv("NHANESData.csv")

# relabelling values for plotting and selecting out NA Data
da["DMDEDUC2x"] = da.DMDEDUC2.replace({1: "<9", 2: "9-11", 3: "HS/GED", 4: "Some college/AA", 5: "College", 
                                       7: "Refused", 9: "Don't know"})
da["DMDMARTLx"] = da.DMDMARTL.replace({1: "Married", 2: "Widowed", 3: "Divorced", 4: "Separated", 5: "Never married",
                                      6: "Living w/partner", 77: "Refused"})
db = da.loc[(da.DMDEDUC2x != "Don't know") & (da.DMDMARTLx != "Refused"), :]

# creating the contingency tables
x=pd.crosstab(db.DMDEDUC2x, db.DMDMARTLx)
x
```

```python
# normalizing contingency 
x.apply(lambda z: z/z.sum(), axis=1)
```

```python
# three way contingency
db.groupby(["RIAGENDR", "DMDEDUC2x", "DMDMARTLx"]).size().unstack().fillna(0).apply(lambda x: x/x.sum(), axis=1)
```

```python
# mosaic plot
# imports
from statsmodels.graphics.mosaicplot import mosaic

# build contingency table
# education vs marital status
x=pd.crosstab(db.DMDEDUC2x, db.DMDMARTLx)
# convert to dict-like Series with a MultiIndex
data=x.stack() 

# plotting 
fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("Proportions")
mosaic(data, gap=0.02, ax=ax, title="Education vs Marital Status")
plt.show()
```
